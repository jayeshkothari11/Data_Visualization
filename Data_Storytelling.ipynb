```python
#Stacked Bar Chart

import pandas as pd
import matplotlib.pyplot as plt

# Read the data from Excel file
data = pd.read_excel('/content/drive/MyDrive/Pycon 2023 Gliwice/Global Web Browser Share.xlsx')

# Set the figure size
plt.figure(figsize=(10 , 6))

# Set the bar width
bar_width = 1


# Get the data for Gold, Silver, and Bronze
Google_Chrome = data['Google Chrome']* 100
Safari = data['Safari']* 100
Internet_Explorer = data['Internet Explorer']* 100
Firefox = data['Firefox']* 100
Opera = data['Opera']* 100
Others = data['Others']* 100


# Get the dates
Date = data['Date']

# Create the bars for all browsers
plt.bar(Date, Google_Chrome,  label='Gold', color='gold')
plt.bar(Date, Safari, bottom=Google_Chrome, label='Silver', color='silver')
plt.bar(Date, Internet_Explorer,  bottom=Google_Chrome+Safari, label='Bronze', color='brown')
plt.bar(Date, Firefox,  bottom=Google_Chrome+Safari+Internet_Explorer, label='Bronze', color='red')
plt.bar(Date, Opera,  bottom=Google_Chrome+Safari+Internet_Explorer+Firefox, label='Bronze', color='green')
plt.bar(Date, Others,  bottom=Google_Chrome+Safari+Internet_Explorer+Firefox+Opera, label='Bronze', color='grey')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('%')
plt.title('Global Web Browser Market Share')

# Get the column names for legends & add Legends
# Remove Date from list of columns
Legend_Data = data.drop(columns=['Date'])
legends = list(Legend_Data.columns)
plt.legend(title='Web Browser', labels=legends, bbox_to_anchor=(1, 1))

# Rotate the x-axis labels for better readability
plt.xticks(rotation=0)

# Add gridlines
plt.grid(True, color='grey', linestyle='dotted')

# Show the plot
plt.show()
```
